```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import skimage.io as io
import os
import scipy.stats as st 
```

CPU

```python
Cellpose_times_path = r"D:\nuclear_seg\Cellpose_preds\Cellpose_times.csv"
StarDist_times_path = r"D:\nuclear_seg\StarDist_preds\StarDist_times.csv"
Mesmer_times_path = r"D:\nuclear_seg\Mesmer_preds\Mesmer_times.csv"

Cellpose_imgs_path = r"D:\nuclear_seg\Cellpose_preds\seg"
StarDist_imgs_path = r"D:\nuclear_seg\StarDist_preds\seg"
Mesmer_imgs_path = r"D:\nuclear_seg\Mesmer_preds\seg"
```

```python
Cellpose_times_df = pd.read_csv(Cellpose_times_path, index_col=0)
Cellpose_times_df['Algorithm'] = 'Cellpose_CPU'
StarDist_times_df = pd.read_csv(StarDist_times_path, index_col=0)
StarDist_times_df['Algorithm'] = 'StarDist_CPU'
Mesmer_times_df = pd.read_csv(Mesmer_times_path, index_col=0)
Mesmer_times_df['Algorithm'] = 'Mesmer_CPU'
```

```python
# Adding the number of nuclei per image

num_nuclei_list = []
for file in StarDist_times_df['img']:
    fpath = os.path.join(StarDist_imgs_path, file)
    num_nuclei = len(np.unique(io.imread(fpath))) - 1
    num_nuclei_list.append(num_nuclei)
StarDist_times_df['num_nuclei'] = num_nuclei_list

num_nuclei_list = []
for file in Mesmer_times_df['img']:
    fpath = os.path.join(Mesmer_imgs_path, file)
    num_nuclei = len(np.unique(io.imread(fpath))) - 1
    num_nuclei_list.append(num_nuclei)
Mesmer_times_df['num_nuclei'] = num_nuclei_list

num_nuclei_list = []
for file in Cellpose_times_df['img']:
    fpath = os.path.join(Cellpose_imgs_path, file)
    num_nuclei = len(np.unique(io.imread(fpath))) - 1
    num_nuclei_list.append(num_nuclei)
Cellpose_times_df['num_nuclei'] = num_nuclei_list
```

```python
num_pixels_list = []
for file in StarDist_times_df['img']:
    fpath = os.path.join(StarDist_imgs_path, file)
    img_arr = io.imread(fpath)
    num_pixels = img_arr.shape[0] * img_arr.shape[1]
    num_pixels_list.append(num_pixels)
StarDist_times_df['num_pixels'] = num_pixels_list

num_pixels_list = []
for file in Mesmer_times_df['img']:
    fpath = os.path.join(Mesmer_imgs_path, file)
    img_arr = io.imread(fpath)
    num_pixels = img_arr.shape[0] * img_arr.shape[1]
    num_pixels_list.append(num_pixels)
Mesmer_times_df['num_pixels'] = num_pixels_list

num_pixels_list = []
for file in Cellpose_times_df['img']:
    fpath = os.path.join(Cellpose_imgs_path, file)
    img_arr = io.imread(fpath)
    num_pixels = img_arr.shape[0] * img_arr.shape[1]
    num_pixels_list.append(num_pixels)
Cellpose_times_df['num_pixels'] = num_pixels_list
```

```python
sns.scatterplot(data=Cellpose_times_df, x='num_nuclei', y='time_seconds', hue='num_pixels')
```

```python
plt.figure(figsize=(20,5))
times_boxplot = sns.boxplot(data=overall_times_df, y='Algorithm', x='time_seconds', palette=['#c03d3e', '#c03d3e', '#845b53', '#845b53', '#d684bd', '#d684bd'], flierprops=dict(marker='d', markerfacecolor='black'))
times_boxplot.set_xlabel(
    xlabel="Computation Time (seconds)", fontweight="bold", fontsize=16
)
times_boxplot.set_ylabel("")
times_boxplot.set_title(
    label="Computation Time for Segmentation Algorithms (27 Fields)", fontweight="bold", fontsize="xx-large"
)
plt.xticks(np.arange(0.0, 301, step=20), fontweight="regular", fontsize=15)
# plt.yticks(fontweight="regular", fontsize=15)
plt.yticks(ticks=[0, 1, 2, 3, 4, 5], labels=["Mesmer (CPU)", "Mesmer (GPU)", "Cellpose (CPU)", "Cellpose (GPU)", "StarDist (CPU)", "StarDist (GPU)"],
           fontweight="bold", fontsize=15)

plt.savefig(r"D:\nuclear_seg\revision_figs\times.png", dpi=500, bbox_inches='tight')
```
